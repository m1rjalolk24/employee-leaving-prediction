Analysis of Correlations
Key Correlations with left (Target Variable)
satisfaction_level (-0.39): There is a moderate negative correlation between satisfaction level and whether an employee left the company. This suggests that employees with lower satisfaction levels are more likely to leave the company.

time_spend_company (0.14): There is a weak positive correlation between time spent at the company and leaving. This may indicate that employees who spend more years at the company have a slightly higher tendency to leave, potentially due to burnout or lack of growth opportunities.

Work_accident (-0.15): There is a weak negative correlation between work accidents and leaving. This suggests that employees who have had work accidents are slightly less likely to leave the company.

average_montly_hours (0.071): There is a very weak positive correlation between average monthly hours and leaving. This suggests that employees who work more hours might be slightly more inclined to leave.

last_evaluation (0.0066) and number_project (0.024): There are almost no correlations between last evaluation, number of projects, and leaving. This suggests these factors do not strongly influence employee retention.

promotion_last_5years (-0.062): There is a very weak negative correlation between promotion in the last five years and leaving. This indicates that employees who have been promoted are slightly less likely to leave.

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.get_dummies(df, columns=['Department', 'salary'], drop_first=True)
X = df.drop(columns=['left'])
y = df['left']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
coefficients = model.coef_[0]
feature_names = X.columns
```

```python
coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
coef_df = coef_df.sort_values(by='Coefficient', ascending=False)
```

```python
coef_df
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

Model is accurate to 78%
